```python
data = load_data('test.csv')
data
```

```python
text_data = preprocess_text(data["text"])
vectorize_with_tfidf(text_data).min()
```

```python
list(map(lambda s: s.split(), text_data))
```

```python
texts = list(map(lambda s: s.split(), text_data))
```

```python
from gensim.test.utils import common_texts
common_texts
```

```python
from gensim.models import Word2Vec

model = Word2Vec(sentences=texts, vector_size=100, window=5, min_count=1, workers=4)
```

```python
vocab, vectors = model.wv.key_to_index, model.wv.vectors
len(vocab), vectors.shape
```

```python
import numpy as np

tweets_vectors = np.array(list(map(
        lambda t: np.array(
            list(map(
                lambda s: vectors[vocab[s]], t))
            ).mean(axis=0),
        texts
    )
))
tweets_vectors.shape
```
